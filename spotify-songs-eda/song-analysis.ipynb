```python
#removing unecessary features, getting song counts, grouping by artist
df_pop = df_pop[['artist', 'bpm', 'nrgy',
       'dnce', 'dB', 'live', 'val', 'dur', 'acous', 'spch', 'pop', 'top year',
       'artist type']]
df_pop['count'] = df.groupby('artist')['artist'].transform('count')
df_pop = df_pop.groupby(['artist']).mean().sort_values('count', ascending = False)
df_pop
```

```python
#interested in how vars relate to pop and count
plt.scatter(df_pop['bpm'], df_pop['count'])
plt.xlabel('bpm')
plt.ylabel('count')
plt.title('Bpm vs number of songs')
```

```python
plt.scatter(df_pop['bpm'], df_pop['pop'])
plt.xlabel('bpm')
plt.ylabel('pop')
plt.title('Bpm vs popularity')
```
