```python
#removing unecessary features, getting song counts, grouping by artist
df_pop = df_pop[['artist', 'bpm', 'nrgy',
       'dnce', 'dB', 'live', 'val', 'dur', 'acous', 'spch', 'pop', 'top year',
       'artist type']]
df_pop_cat = df_pop
df_pop['count'] = df.groupby('artist')['artist'].transform('count')
df_pop = df_pop.groupby('artist').mean().sort_values('count', ascending = False)
df_pop.head()
```

```python
numerical_var = ['bpm', 'nrgy',
       'dnce', 'dB', 'live', 'val', 'dur', 'acous', 'spch', 'top year']
#Graphs of numerical vars vs count
for i in numerical_var:
    plt.scatter(df_pop[i], df_pop['count'])
    plt.xlabel(i)
    plt.ylabel('count')
    title = i + ' vs count'
    plt.title(title)
    plt.show()
```

```python
#Graphs of numerical vars vs popularity
for i in numerical_var:
    plt.scatter(df_pop[i], df_pop['pop'])
    plt.xlabel(i)
    plt.ylabel('popularity')
    title = i + ' vs pop'
    plt.title(title)
    plt.show()
```
